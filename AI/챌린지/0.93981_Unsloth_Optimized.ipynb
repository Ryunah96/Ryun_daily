# 3. ë¼ì´ë¸ŒëŸ¬ë¦¬, ë°ì´í„°, ì„¤ì • ë¡œë“œ

```python
import os, random
import pandas as pd
from PIL import Image
import torch
from datasets import Dataset, DatasetDict
from unsloth import FastVisionModel
from trl import SFTTrainer, SFTConfig
from unsloth.chat_templates import get_chat_template
from unsloth.trainer import UnslothVisionDataCollator
from tqdm import tqdm

# ì´ë¯¸ì§€ ë¡œë“œ ì‹œ í”½ì…€ ì œí•œ í•´ì œ
Image.MAX_IMAGE_PIXELS = None

# ë””ë°”ì´ìŠ¤ GPU ìš°ì„  ì‚¬ìš© ì„¤ì •
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# ì„¤ì •ê°’ ì •ì˜
MODEL_ID = "unsloth/Qwen3-VL-8B-Instruct-unsloth-bnb-4bit" # Unslothì˜ 4bit ì–‘ìí™” ëª¨ë¸
MAX_SEQ_LENGTH = 2048
MAX_NEW_TOKENS = 8
SEED = 42
random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ë°ì´í„°ì…‹ ë¡œë“œ
train_df = pd.read_csv("/content/train.csv")
test_df  = pd.read_csv("/content/test.csv")

# í•™ìŠµ ì‹œê°„ì„ ì¤„ì´ê¸° ìœ„í•´ 200ê°œ ìƒ˜í”Œë§Œ ì‚¬ìš© (ì‹¤ì œ ëŒ€íšŒì—ì„œëŠ” ì „ì²´ ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”)
train_df = train_df.sample(n=200, random_state=SEED).reset_index(drop=True)
```

# 4. Unsloth ëª¨ë¸ ë° LoRA ì–´ëŒ‘í„° ë¡œë”©

Unslothì˜ `FastVisionModel`ì„ ì‚¬ìš©í•˜ì—¬ 4bit ì–‘ìí™”ëœ ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ ê°„í¸í•˜ê²Œ ë¡œë“œí•©ë‹ˆë‹¤. ì´í›„ LoRA ì–´ëŒ‘í„°ë¥¼ ì¶”ê°€í•˜ì—¬ í•™ìŠµ ì¤€ë¹„ë¥¼ ë§ˆì¹©ë‹ˆë‹¤.

```python
model, tokenizer = FastVisionModel.from_pretrained(
    model_name = MODEL_ID,
    max_seq_length = MAX_SEQ_LENGTH,
    load_in_4bit = True, # False for LoRA 16bit
    fast_inference = False, # Enable vLLM fast inference
    gpu_memory_utilization = 0.8, # Reduce if out of memory
)

# LoRA ì–´ëŒ‘í„° ì¶”ê°€
model = FastVisionModel.get_peft_model(
    model,
    finetune_vision_layers     = True,  # False if not finetuning vision layers
    finetune_language_layers   = True,  # False if not finetuning language layers
    finetune_attention_modules = True,  # False if not finetuning attention layers
    finetune_mlp_modules       = True,  # False if not finetuning MLP layers

    r = 32,           # The larger, the higher the accuracy, but might overfit
    lora_alpha = 32,  # Recommended alpha == r at least
    lora_dropout = 0.05,
    bias = "none",
    random_state = SEED,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
    use_gradient_checkpointing = "unsloth", # Reduces memory usage
    # target_modules = "all-linear", # Optional now! Can specify a list if needed
)

model.print_trainable_parameters()
```

```python
from trl.trainer.sft_trainer import DataCollatorForVisionLanguageModeling
from transformers import AutoProcessor

processor = AutoProcessor.from_pretrained(MODEL_ID)
collator = DataCollatorForVisionLanguageModeling(processor)
```

# 5. í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ ë° ë°ì´í„° í¬ë§·íŒ…

í•™ìŠµ ë° ì¶”ë¡ ì— ì‚¬ìš©í•  í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ì„ ì •ì˜í•˜ê³ , Hugging Face `datasets` ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ëª¨ë¸ì˜ Chat Templateì— ë§ê²Œ ë³€í™˜í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì€ ê¸°ì¡´ì˜ Custom Dataset/Dataloaderë¥¼ ëŒ€ì²´í•©ë‹ˆë‹¤.

```python
# Pandas DataFrameì„ Hugging Face Datasetìœ¼ë¡œ ë³€í™˜
raw_dataset = Dataset.from_pandas(train_df)

# Resize to (512, 512) and handle image loading errors
def convert_to_rgb(example):
    try:
        example["decoded_image"] = Image.open(example["path"]).resize((512, 512)).convert("RGB")
    except Exception as e:
        print(f"Error loading image {example['path']}: {e}")
        # Create a white dummy image
        example["decoded_image"] = Image.new("RGB", (512, 512), color = 'white')
    return example

raw_dataset = raw_dataset.map(convert_to_rgb)
```

```python
# ì‹œìŠ¤í…œ í”„ë¡¬í¬íŠ¸ë„ ì»¤ìŠ¤í„°ë§ˆì´ì§• í•„ìˆ˜! (ì •í™•ë¥  5~15% í–¥ìƒ)
SYSTEM_INSTRUCT = (
    "You are a helpful visual question answering assistant. "
    "Answer using exactly one letter among a, b, c, or d. No explanation."
)

def build_mc_prompt(question, a, b, c, d):
    return (
        f"{question}\n"
        f"(a) {a}\n(b) {b}\n(c) {c}\n(d) {d}\n\n"
        "ì •ë‹µì„ ë°˜ë“œì‹œ a, b, c, d ì¤‘ í•˜ë‚˜ì˜ ì†Œë¬¸ì í•œ ê¸€ìë¡œë§Œ ì¶œë ¥í•˜ì„¸ìš”."
    )

# í•™ìŠµ ë°ì´í„°ë¥¼ ëª¨ë¸ì˜ ëŒ€í™” í˜•ì‹(messages)ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def make_conversation(example):
    user_text = build_mc_prompt(str(example["question"]), str(example["a"]), str(example["b"]), str(example["c"]), str(example["d"]))
    gold = str(example["answer"]).strip().lower()

    messages = [
        {"role":"system","content":[{"type":"text","text":SYSTEM_INSTRUCT}]},
        {
            "role":"user",
            "content":[
                {"type":"image"}, # Placeholder for the image
                {"type":"text","text":user_text}  # The text part of the prompt
                ]
         },
          {
              "role":"assistant",
              "content":[
                  {"type":"text","text":gold}
                  ]
            }
    ]
    # The actual image data is kept separate for the processor
    return {"messages": messages, "images": [example["decoded_image"]]}

dataset = raw_dataset.map(make_conversation, remove_columns=["path", "a", "b", "c", "d", "question", "id", "answer", "decoded_image"])

formatted_dataset = dataset

# í›ˆë ¨/ê²€ì¦ ë°ì´í„° ë¶„ë¦¬
dataset_split = formatted_dataset.train_test_split(test_size=0.1, seed=SEED)
train_dataset = dataset_split["train"]
valid_dataset = dataset_split["test"]

print("Final training sample:\n", train_dataset[0])
```

# 6. SFTTrainerë¥¼ ì‚¬ìš©í•œ íŒŒì¸íŠœë‹

Unslothì™€ `trl` ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ `SFTTrainer`ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ íŒŒì¸íŠœë‹ì„ ì§„í–‰í•©ë‹ˆë‹¤. ì´ ë°©ì‹ì€ ê¸°ì¡´ì˜ ìˆ˜ë™ í•™ìŠµ ë£¨í”„ë³´ë‹¤ í›¨ì”¬ ê°„ê²°í•˜ê³  íš¨ìœ¨ì ì…ë‹ˆë‹¤.

ê³µì‹ docs : https://huggingface.co/docs/trl/main/sft_trainer

- 200ê°œ ìƒ˜í”Œ í•™ìŠµ ì‹œ ì•½ 5~10ë¶„ ì†Œìš”ë©ë‹ˆë‹¤.

```python
# --- 3. ëª¨ë¸ í•™ìŠµ ---

# ëª¨ë¸ì„ í•™ìŠµ ëª¨ë“œë¡œ í™œì„±í™”
FastVisionModel.for_training(model)

# SFTTrainer ì„¤ì •
trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    data_collator=collator,
    args=SFTConfig(
        per_device_train_batch_size=2,         # ğŸ”¼ ë°°ì¹˜ í¬ê¸°ë¥¼ 2ë¡œ (GPU ì—¬ìœ ê°€ ìˆë‹¤ë©´)
        per_device_eval_batch_size=2,
        gradient_accumulation_steps=8,         # ğŸ”¼ ë” ìì£¼ ì—…ë°ì´íŠ¸í•˜ì—¬ ì•ˆì •ì  í•™ìŠµ
        warmup_ratio=0.1,                     # ğŸ”¼ ì „ì²´ ìŠ¤í…ì˜ 5%ë¥¼ ì›Œë°ì—…
        max_steps=300,                         # ğŸ”¼ í•™ìŠµ ìŠ¤í…ì„ ëŠ˜ë ¤ ë” ì¶©ë¶„í•œ ìˆ˜ë ´ ìœ ë„
        learning_rate=3e-5,                    # ğŸ”½ ì¡°ê¸ˆ ë” ì•ˆì •ì ì¸ í•™ìŠµë¥ 
        max_grad_norm=1.0,                     # ğŸ”¼ gradient clippingìœ¼ë¡œ í­ì£¼ ë°©ì§€
        logging_steps=2,
        eval_strategy="steps",
        eval_steps=20,                         # ğŸ”¼ í‰ê°€ ì£¼ê¸° ëŠ˜ë ¤ í•™ìŠµ ì‹œê°„ ì ˆì•½
        save_strategy="steps",                 # ğŸ”¼ ì£¼ê¸°ì ìœ¼ë¡œ ì²´í¬í¬ì¸íŠ¸ ì €ì¥
        save_steps=20,
        save_total_limit=3,                    # ğŸ”¼ ìµœê·¼ 3ê°œë§Œ ì €ì¥ (ë””ìŠ¤í¬ ì ˆì•½)
        optim="paged_adamw_8bit",              # ğŸ”¼ ë” íš¨ìœ¨ì ì¸ ì˜µí‹°ë§ˆì´ì €
        weight_decay=0.05,                     # ğŸ”¼ ì•½ê°„ì˜ ê°€ì¤‘ì¹˜ ê°ì‡ ë¡œ ì¼ë°˜í™” í–¥ìƒ
        lr_scheduler_type="cosine",            # ğŸ”¼ ì½”ì‚¬ì¸ ìŠ¤ì¼€ì¤„ëŸ¬ë¡œ ë¶€ë“œëŸ¬ìš´ decay
        report_to="none",
        seed=SEED,
        output_dir="outputs",
        remove_unused_columns=False,
        dataset_kwargs={"skip_prepare_dataset": True},
        max_length=MAX_SEQ_LENGTH,
        gradient_checkpointing=True,           # ğŸ”¼ GPU ë©”ëª¨ë¦¬ ì ˆì•½
        fp16=True,                             # ğŸ”¼ í˜¼í•© ì •ë°€ë„ë¡œ í•™ìŠµ ì†ë„ í–¥ìƒ
        push_to_hub=False,                     # HuggingFace ì—…ë¡œë“œ ë¹„í™œì„±í™”
        load_best_model_at_end=True,           # ğŸ”¼ ê°€ì¥ ì¢‹ì€ eval ê²°ê³¼ ëª¨ë¸ ìë™ ë¡œë“œ
        metric_for_best_model="eval_loss",     # ğŸ”¼ ê¸°ì¤€ ë©”íŠ¸ë¦­ ì„¤ì •
        greater_is_better=False,               # ì†ì‹¤ ê¸°ì¤€ì´ë©´ False
    ),
)

# í•™ìŠµ ì‹œì‘
print("ğŸš€ Starting fine-tuning...")
trainer.train()
print("âœ… Training finished.")
```

# 7. ì¶”ë¡  ë° ì œì¶œ íŒŒì¼ ìƒì„±

í•™ìŠµëœ LoRA ì–´ëŒ‘í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì¶”ë¡ ì„ ìˆ˜í–‰í•˜ê³ , `submission.csv` íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.

- ì „ì²´ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì¶”ë¡  ì‹œ ì•½ 30ë¶„~1ì‹œê°„ ì†Œìš”ë©ë‹ˆë‹¤.

```python
FastVisionModel.for_inference(model) # ì¶”ë¡  ëª¨ë“œ í™œì„±í™”

# ëª¨ë¸ì˜ ì‘ë‹µì—ì„œ ì •ë‹µ(a, b, c, d)ë§Œ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜
def extract_choice(text: str) -> str:
    text = text.strip().lower()
    lines = [l.strip() for l in text.splitlines() if l.strip()]
    if not lines:
        return "a" # ì‘ë‹µì´ ì—†ëŠ” ê²½ìš° 'a'ë¡œ ëŒ€ì²´
    last = lines[-1]
    if last in ["a", "b", "c", "d"]:
        return last
    tokens = last.split()
    for tok in tokens:
        if tok in ["a", "b", "c", "d"]:
            return tok
    return "a" # ì •ë‹µì„ ì°¾ì§€ ëª»í•œ ê²½ìš° 'a'ë¡œ ëŒ€ì²´

# ì¶”ë¡  ë£¨í”„
preds = []
for i in tqdm(range(len(test_df))):
    row = test_df.iloc[i]
    image = Image.open(row["path"]).convert("RGB")
    user_text = build_mc_prompt(row["question"], row["a"], row["b"], row["c"], row["d"])

    # Updated prompt structure to correctly pass text
    messages = [
        {"role":"system","content":[{"type":"text","text":SYSTEM_INSTRUCT}]},
        {
            "role":"user",
            "content":[
                {"type":"image"}, # Placeholder for the image
                {"type":"text","text":user_text}  # The text part of the prompt
                ]
         }
      ]

    prompt = tokenizer.apply_chat_template(
            messages,
            tokenize = False,
            add_generation_prompt = True, # Must add assistant
    )

    # Chat Template ì ìš© - Pass the list of dictionaries directly to the tokenizer
    # Corrected tokenizer call to pass the text and image separately
    inputs = tokenizer(
        image,
        prompt,
        add_special_tokens = False,
        return_tensors = "pt",
    ).to("cuda")


    outputs = model.generate(
        **inputs,
        max_new_tokens=MAX_NEW_TOKENS,
        # use_cache=True
    )

    # skip_special_tokens=Trueë¡œ íŠ¹ìˆ˜ í† í° ì œê±°
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    preds.append({"id": row["id"], "answer": extract_choice(text)})

# ì œì¶œ íŒŒì¼ ìƒì„±
submission = pd.DataFrame(preds, columns=['id', 'answer'])
submission.to_csv("/content/submission.csv", index=False)
print("Saved /content/submission.csv")
print("Submission file sample:")
print(submission.head())
```
